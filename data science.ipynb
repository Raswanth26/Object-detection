```python
import cv2
import numpy as np
import os

# Path to the video file
video_path = r"D:\yolo project\yolo input.mp4.webm"
# Path to save cropped objects
output_folder = r"D:\yolo project\output_folder"
# Create output folder if it doesn't exist
os.makedirs(output_folder, exist_ok=True)

# Load YOLO
net = cv2.dnn.readNet(r"C:\Users\jayapriya\Downloads\yolov4.weights", r"C:\Users\jayapriya\Downloads\yolov4.cfg")
layer_names = net.getLayerNames()
output_layers = [layer_names[i - 1] for i in net.getUnconnectedOutLayers()]
classes = []
with open(r"C:\Users\jayapriya\Downloads\coco.names", "r") as f:
    classes = [line.strip() for line in f.readlines()]

# Open the video file
cap = cv2.VideoCapture(video_path)
frame_id = 0

while True:
    ret, frame = cap.read()
    if not ret:
        break

    height, width, channels = frame.shape

    # Resize the frame to improve object detection
    blob = cv2.dnn.blobFromImage(frame, 0.00392, (416, 416), (0, 0, 0), True, crop=False)
    net.setInput(blob)
    outs = net.forward(output_layers)

    # Showing information on the screen and getting the confidence score of the algorithm
    class_ids = []
    confidences = []
    boxes = []
    for out in outs:
        for detection in out:
            scores = detection[5:]
            class_id = np.argmax(scores)
            confidence = scores[class_id]
            if confidence > 0.3:  # Adjusted threshold value
                # Object detected
                center_x = int(detection[0] * width)
                center_y = int(detection[1] * height)
                w = int(detection[2] * width)
                h = int(detection[3] * height)

                # Rectangle coordinates
                x = int(center_x - w / 2)
                y = int(center_y - h / 2)

                boxes.append([x, y, w, h])
                confidences.append(float(confidence))
                class_ids.append(class_id)

    indexes = cv2.dnn.NMSBoxes(boxes, confidences, 0.5, 0.4)

    if len(indexes) > 0:
        print(f"Detected {len(indexes)} objects in frame {frame_id}")
    else:
        print(f"No objects detected in frame {frame_id}")

    for i in range(len(boxes)):
        if i in indexes:
            x, y, w, h = boxes[i]
            label = str(classes[class_ids[i]])
            confidence = confidences[i]
            color = (0, 255, 0)

            print(f"Object {label} detected at ({x}, {y}) with confidence {confidence:.2f}")

            # Draw bounding box on the original frame
            cv2.rectangle(frame, (x, y), (x+w, y+h), color, 2)

            # Crop the detected object
            crop_img = frame[max(0, y):min(y+h, height), max(0, x):min(x+w, width)]

            # Save the cropped object
            crop_img_path = os.path.join(output_folder, f"{frame_id}_{label}_{confidence:.2f}.jpg")
            try:
                cv2.imwrite(crop_img_path, crop_img)
                print(f"Saved cropped image to {crop_img_path}")
            except Exception as e:
                print(f"Error writing image to {crop_img_path}: {str(e)}")

    # Display the original frame with bounding boxes
    cv2.imshow("Frame", frame)
    cv2.waitKey(1)

    frame_id += 20

cap.release()
cv2.destroyAllWindows()
```
